# Renombrado y Combinación

En este JNB se revisará el tema de Renombrado y Combinación en Pandas.

```python
import pandas as pd
reviews = pd.read_csv("https://raw.githubusercontent.com/GTD2401-ACD-GA-EJ26/ACD.datasets.kaggle/refs/heads/main/winemag-data-130k-v2.txt", index_col=0)
reviews
```

La primera función es rename(), la cual te permite cambiar los nombres de los indices y/o de las columnas. Por ejemplo, para cambiar la columna points en nuestro conjunto de datos a score, hariamos lo siguiente:

```python
reviews.rename(columns={"points": "score"})
```

rename() te permite renombrar valores del índice o de las columnas especificando los parámetros index o columns, respectivamente. Admite varios formatos de entrada, pero usualmente un diccionario de python es lo más conveniente. Aquí hay un ejemplo usándolo para renombrar algunos elementeos del índice:

```python
reviews.rename(index={0: "firstEntry", 1: "secondEntry"})
```

Es problabe que se renombren columnas con frecuencia, pero los valores del índice muy raramente. para esto último, set_index() suele ser más práctico.

Tanto el indice de las filas como el de las columnas pueden tener su propio atributo name. El método complementario rename_axis() puede utilizarse para cambiar estos nombres. Por ejemplo:

```python
reviews.rename_axis("wines", axis="rows").rename_axis("fields", axis="columns")
```

*Combinar (combining)*
Al realizar operacioes en conjunto de datos, a veces necesitaremos combinar diferentes DataFrames y/o Series de formas no triviales. Pandas tiene tres métodos principales para hacer esto. En orden de complejidad creciente, estos son: concat(), join() y merge(). La mayor parte de lo que puede hacer merge() tambien se puede hacer de forma más sencilla con join(), por lo que la omitiremos y nos centraremos en las dos primeras

```python
canadian_youtube = pd.read_csv("https://raw.githubusercontent.com/GTD2401-ACD-GA-EJ26/ACD.datasets.kaggle/refs/heads/main/CAvideos.txt", index_col=0)
british_youtube = pd.read_csv("https://raw.githubusercontent.com/GTD2401-ACD-GA-EJ26/ACD.datasets.kaggle/refs/heads/main/GBvideos.txt", index_col=0)
```

*Concat()*
El método de combinación más simple es concat(). Dada una lista de elementos, esta función "pegará" esos elementos a lo largo de un eje.

Esto es útil cuando tenemos datos en diferentes objetos DataFrame o Series que tienen los mismos campos (columnas). Por ejemplo, un conjunto de datos de videos de youtube que divide los datos según el país de origen (canada y reino unido). Si queremos estudiar varios países simultáneamente, podemos usar concat() para unirlos

```python
pd.concat([canadian_youtube, british_youtube])
```

*join()*
El combinador intermedio en términos de complejidad es join(). Este te permite combinar diferentes objetos DataFrame que tienen un índie en común. Por ejemplo, para obtener videos que resultaron ser tendencia el mismo día tanto en Canadá como en el Reino Unido, podríamos hacer lo siguiente:

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix="_CAN", rsuffix="_UK")
```
